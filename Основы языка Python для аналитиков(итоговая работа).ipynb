```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("kc-house-data.csv")
df.head()
```

```
df = pd.read_csv("laptop_price.csv")
df.head()
```

```
plt.figure(figsize=(8, 6))

sns.histplot(df['price'], bins=60)

plt.title('Стоимость недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Количество');
```

* Вывод: Основное колличество добов стоимостью до 1 млн.



```
plt.figure(figsize=(8, 6))
sns.histplot(df['sqft_living'], bins=60)
plt.title('Распределение жилой квадратуры')
plt.xlabel('Квадратные футы')
plt.ylabel('Количество');
```

* Вывод: основное количество проданных домов до 3000 кв.м



```
plt.figure(figsize=(8, 6))
sns.histplot(df['yr_built'], bins=60)
plt.title('Распределение года постройки')
plt.xlabel('Годы постройки жилья')
plt.ylabel('Количество');
```

* Вывод: основное количество домов построено с 1944 по 2007 года. 1967 и 1986 году был спад. Меньше всего построено домов с 1900 по 1943 года.
